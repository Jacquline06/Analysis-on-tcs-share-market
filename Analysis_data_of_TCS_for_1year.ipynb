```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters =3)
```

```python
model.fit(data.iloc[:,:-2])
```

```python
group=model.predict(data.iloc[:,:-2])
```

```python
wcss=[]
for i in range(1,247):
  model=KMeans(n_clusters=i)
  model.fit(data)
  wcss.append(model.inertia_)
```
